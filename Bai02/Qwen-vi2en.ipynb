```python
# ============================================================================
# CELL 1: CÃ i Ä‘áº·t thÆ° viá»‡n cáº§n thiáº¿t
# ============================================================================
import subprocess
import sys

# CÃ i Ä‘áº·t cÃ¡c packages chÃ­nh
subprocess.check_call([
    sys.executable, "-m", "pip", "install", "-q", "-U",
    "transformers>=4.51.0", "accelerate>=1.0.0", "trl>=0.9.6", "datasets>=2.18.0"
])

# CÃ i Ä‘áº·t Unsloth vÃ  bitsandbytes
subprocess.check_call([
    sys.executable, "-m", "pip", "install", "-q",
    "unsloth", "bitsandbytes"
])

print("âœ“ ÄÃ£ cÃ i Ä‘áº·t cÃ¡c thÆ° viá»‡n cáº§n thiáº¿t")
```

```python
# ============================================================================
# CELL 2: Cáº¥u hÃ¬nh vÃ  load mÃ´ hÃ¬nh Qwen3 1.7B vá»›i Unsloth (4-bit quantization)
# ============================================================================

import os
from unsloth import FastLanguageModel
import torch

# === Cáº¥u hÃ¬nh Ä‘Æ°á»ng dáº«n dataset MedicalDataset_VLSP ===
# Tá»± Ä‘á»™ng tÃ¬m dataset trÃªn Kaggle
dataset_paths = [
    "/kaggle/input/medicaldataset-vlsp/MedicalDataset_VLSP",
    "/kaggle/input/medicaldataset-vlsp",
    "MedicalDataset_VLSP",
    "./MedicalDataset_VLSP"
]

DATA_ROOT = None
for path in dataset_paths:
    train_en_file = os.path.join(path, "train.en.txt")
    if os.path.exists(path) and os.path.exists(train_en_file):
        DATA_ROOT = path
        print(f"âœ“ TÃ¬m tháº¥y dataset táº¡i: {DATA_ROOT}")
        break

if DATA_ROOT is None:
    raise FileNotFoundError(
        "KhÃ´ng tÃ¬m tháº¥y dataset MedicalDataset_VLSP!\n"
        "ÄÃ£ thá»­ cÃ¡c Ä‘Æ°á»ng dáº«n:\n" + "\n".join(f"  - {p}" for p in dataset_paths)
    )

TRAIN_EN_FILE = os.path.join(DATA_ROOT, "train.en.txt")
TRAIN_VI_FILE = os.path.join(DATA_ROOT, "train.vi.txt")

# Cáº¥u hÃ¬nh sá»‘ lÆ°á»£ng samples (None = dÃ¹ng táº¥t cáº£)
LIMIT_SAMPLES = None  # CÃ³ thá»ƒ set sá»‘ nhá» (vd: 20000) Ä‘á»ƒ test nhanh

# === Cáº¥u hÃ¬nh mÃ´ hÃ¬nh ===
BASE_MODEL = "unsloth/Qwen3-1.7B-unsloth-bnb-4bit"
MAX_SEQUENCE_LENGTH = 2048
RANDOM_SEED = 3407

# Set random seed Ä‘á»ƒ Ä‘áº£m báº£o reproducibility
torch.manual_seed(RANDOM_SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(RANDOM_SEED)

print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")

# Load Qwen3 1.7B vá»›i 4-bit quantization qua Unsloth
print(f"Äang load mÃ´ hÃ¬nh: {BASE_MODEL}")
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=BASE_MODEL,
    max_seq_length=MAX_SEQUENCE_LENGTH,
    load_in_4bit=True,      # Sá»­ dá»¥ng QLoRA 4-bit Ä‘á»ƒ tiáº¿t kiá»‡m memory
    load_in_8bit=False,
    full_finetuning=False, # Chá»‰ fine-tune vá»›i LoRA, khÃ´ng full fine-tune
    # token="hf_...",      # ThÃªm HuggingFace token náº¿u model lÃ  gated
)

print("âœ“ ÄÃ£ load mÃ´ hÃ¬nh thÃ nh cÃ´ng")
```

```python
# ============================================================================
# CELL 3: Cáº¥u hÃ¬nh LoRA theo API Unsloth
# ============================================================================

print("Äang cáº¥u hÃ¬nh LoRA...")

model = FastLanguageModel.get_peft_model(
    model,
    r=32,  # LoRA rank (theo khuyáº¿n nghá»‹ cá»§a Unsloth)
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ],
    lora_alpha=32,
    lora_dropout=0.0,
    bias="none",
    use_gradient_checkpointing="unsloth",  # Sá»­ dá»¥ng gradient checkpointing cá»§a Unsloth
    random_state=RANDOM_SEED,
    use_rslora=False,
    loftq_config=None,
)

# Kiá»ƒm tra device cá»§a model
first_param = next(model.parameters())
print(f"âœ“ Model device: {first_param.device}")
print("âœ“ LoRA Ä‘Ã£ Ä‘Æ°á»£c cáº¥u hÃ¬nh")
```

```python
# ============================================================================
# CELL 4: Load dá»¯ liá»‡u song ngá»¯(VI -> EN)
# ============================================================================

from datasets import Dataset
import pandas as pd

def load_bilingual_data(vi_file, en_file, max_samples=None):
    """Load dá»¯ liá»‡u song ngá»¯ tá»« file VI vÃ  EN (VI -> EN)"""
    with open(vi_file, "r", encoding="utf-8") as f:
        vietnamese_lines = [line.rstrip("\n") for line in f]
    
    with open(en_file, "r", encoding="utf-8") as f:
        english_lines = [line.rstrip("\n") for line in f]

    if len(vietnamese_lines) != len(english_lines):
        raise ValueError(
            f"Sá»‘ dÃ²ng khÃ´ng khá»›p! VI: {len(vietnamese_lines)}, EN: {len(english_lines)}"
        )

    total = len(vietnamese_lines)
    if max_samples is not None and max_samples > 0:
        total = min(total, max_samples)

    vietnamese_lines = vietnamese_lines[:total]
    english_lines = english_lines[:total]

    print(f"âœ“ ÄÃ£ load {total} cáº·p dá»¯ liá»‡u song ngá»¯ (VI -> EN)")
    return vietnamese_lines, english_lines

# Load dá»¯ liá»‡u training (VI -> EN)
vietnamese_sentences, english_sentences = load_bilingual_data(
    TRAIN_VI_FILE, TRAIN_EN_FILE, LIMIT_SAMPLES
)

# Äá»‹nh nghÄ©a system prompt cho task dá»‹ch mÃ¡y (VI -> EN)
TRANSLATION_SYSTEM_PROMPT = (
    "Báº¡n lÃ  há»‡ thá»‘ng dá»‹ch mÃ¡y chuyÃªn nghiá»‡p, dá»‹ch tá»« tiáº¿ng Viá»‡t sang tiáº¿ng Anh "
    "trong lÄ©nh vá»±c y há»c / y sinh. LuÃ´n giá»¯ Ä‘Ãºng Ã½ nghÄ©a, thuáº­t ngá»¯ y khoa vÃ  "
    "khÃ´ng thÃªm báº¥t ká»³ giáº£i thÃ­ch hay ná»™i dung phá»¥ nÃ o."
)

def create_chat_format(vietnamese_text, english_text):
    """Táº¡o format chat cho má»™t cáº·p cÃ¢u dá»‹ch (VI -> EN)"""
    # Má»—i sample lÃ  má»™t Ä‘oáº¡n há»™i thoáº¡i vá»›i system, user vÃ  assistant
    conversation = [
        {"role": "system", "content": TRANSLATION_SYSTEM_PROMPT},
        {"role": "user", "content": vietnamese_text},   # CÃ¢u nguá»“n (tiáº¿ng Viá»‡t)
        {"role": "assistant", "content": english_text}, # CÃ¢u Ä‘Ã­ch (tiáº¿ng Anh)
    ]
    
    # Ãp dá»¥ng chat template cá»§a Qwen3 (táº¯t thinking mode cho task MT)
    formatted_text = tokenizer.apply_chat_template(
        conversation,
        tokenize=False,
        add_generation_prompt=False,
        enable_thinking=False,  # Táº¯t thinking mode cho translation task
    )
    return formatted_text

# Táº¡o dataset tá»« cÃ¡c cáº·p cÃ¢u (VI -> EN)
training_texts = [
    create_chat_format(vi, en)
    for vi, en in zip(vietnamese_sentences, english_sentences)
]

# Chuyá»ƒn Ä‘á»•i sang Dataset format
dataframe = pd.DataFrame({"text": training_texts})
training_dataset = Dataset.from_pandas(dataframe)
training_dataset = training_dataset.shuffle(seed=RANDOM_SEED)

# Hiá»ƒn thá»‹ sample Ä‘áº§u tiÃªn
sample_text = training_dataset[0]["text"]
print(f"\nSample text (500 kÃ½ tá»± Ä‘áº§u):\n{sample_text[:500]}")
print(f"\nâœ“ Tá»•ng sá»‘ máº«u training: {len(training_dataset)}")
```

```python
# ============================================================================
# CELL 5: Fine-tune mÃ´ hÃ¬nh Qwen3 báº±ng SFTTrainer (theo style Unsloth)
# ============================================================================

from trl import SFTTrainer, SFTConfig

print("\nÄang khá»Ÿi táº¡o trainer...")

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=training_dataset,
    eval_dataset=None,  # CÃ³ thá»ƒ thÃªm validation set sau náº¿u cáº§n
    args=SFTConfig(
        dataset_text_field="text",  # TÃªn cá»™t chá»©a text Ä‘á»ƒ train
        per_device_train_batch_size=2,  # Batch size phÃ¹ há»£p vá»›i T4 16GB
        gradient_accumulation_steps=4,  # Effective batch size = 2 * 4 = 8
        warmup_steps=5,
        # num_train_epochs=1,  # CÃ³ thá»ƒ set theo epoch náº¿u muá»‘n
        max_steps=4000,  # Sá»‘ steps Ä‘á»ƒ train (cÃ³ thá»ƒ Ä‘iá»u chá»‰nh)
        learning_rate=2e-4,  # Learning rate theo khuyáº¿n nghá»‹ Unsloth
        logging_steps=10,
        optim="adamw_8bit",  # Optimizer 8-bit Ä‘á»ƒ tiáº¿t kiá»‡m memory
        weight_decay=0.001,
        lr_scheduler_type="linear",
        seed=RANDOM_SEED,
        report_to="none",  # Táº¯t logging lÃªn WandB/TensorBoard
    ),
)

print("âœ“ Trainer Ä‘Ã£ Ä‘Æ°á»£c khá»Ÿi táº¡o")
print("\n" + "="*60)
print("Báº®T Äáº¦U TRAINING")
print("="*60)

training_stats = trainer.train()

print("\n" + "="*60)
print("TRAINING HOÃ€N Táº¤T")
print("="*60)
print(f"âœ“ Tá»•ng sá»‘ global steps: {trainer.state.global_step}")
```

```python
# ============================================================================
# CELL 6: LÆ°u LoRA adapter vÃ  tokenizer
# ============================================================================

SAVE_DIR = "qwen3-1_7b-vi-en-mt"  # ThÆ° má»¥c lÆ°u káº¿t quáº£ (VI -> EN)
os.makedirs(SAVE_DIR, exist_ok=True)

lora_adapter_path = os.path.join(SAVE_DIR, "lora_adapter")
tokenizer_path = os.path.join(SAVE_DIR, "tokenizer")

model.save_pretrained(lora_adapter_path)
tokenizer.save_pretrained(tokenizer_path)

print(f"\nâœ“ ÄÃ£ lÆ°u LoRA adapter táº¡i: {lora_adapter_path}")
print(f"âœ“ ÄÃ£ lÆ°u tokenizer táº¡i: {tokenizer_path}")
```

```python
# ============================================================================
# CELL 7: Test nhanh dá»‹ch vá»›i mÃ´ hÃ¬nh Ä‘Ã£ fine-tune (VI -> EN)
# ============================================================================

from unsloth import FastLanguageModel
import torch

# Chuyá»ƒn model sang inference mode
FastLanguageModel.for_inference(model)

def translate_vietnamese_to_english(vietnamese_text, max_new_tokens=128):
    """Dá»‹ch má»™t cÃ¢u tiáº¿ng Viá»‡t sang tiáº¿ng Anh"""
    chat_messages = [
        {"role": "system", "content": TRANSLATION_SYSTEM_PROMPT},
        {"role": "user", "content": vietnamese_text},
    ]
    
    # Táº¡o prompt vá»›i generation prompt
    prompt_text = tokenizer.apply_chat_template(
        chat_messages,
        tokenize=False,
        add_generation_prompt=True,
        enable_thinking=False,
    )
    
    # Tokenize vÃ  chuyá»ƒn sang device
    inputs = tokenizer(prompt_text, return_tensors="pt").to(model.device)

    # Generate translation
    with torch.no_grad():
        generated = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,  # Greedy decoding cho translation
        )

    # Extract pháº§n Ä‘Æ°á»£c generate (bá» qua pháº§n prompt)
    prompt_length = inputs["input_ids"].shape[1]
    generated_ids = generated[0][prompt_length:]
    
    # Decode vÃ  tráº£ vá»
    translation = tokenizer.decode(generated_ids, skip_special_tokens=True).strip()
    return translation

# Test vá»›i 3 cÃ¢u Ä‘áº§u
print("\n" + "="*60)
print("TEST Dá»ŠCH (3 cÃ¢u Ä‘áº§u) - VI -> EN")
print("="*60)
for idx in range(3):
    source = vietnamese_sentences[idx]
    reference = english_sentences[idx]
    prediction = translate_vietnamese_to_english(source)
    
    print(f"\n[Sample {idx+1}]")
    print(f"VI  : {source}")
    print(f"REF : {reference}")
    print(f"PRED: {prediction}")
    print("-" * 80)
```

```python
# ============================================================================
# CELL 8: CÃ i Ä‘áº·t thÆ° viá»‡n Ä‘Ã¡nh giÃ¡ vÃ  tÃ­nh SacreBLEU trÃªn public_test (VI -> EN)
# ============================================================================

import subprocess
subprocess.check_call([
    sys.executable, "-m", "pip", "install", "-q",
    "sacrebleu", "evaluate"
])

import os
import torch
from tqdm.auto import tqdm
import evaluate
from unsloth import FastLanguageModel

# === ÄÆ°á»ng dáº«n test set (VI -> EN) ===
TEST_VI_FILE = os.path.join(DATA_ROOT, "public_test.vi.txt")
TEST_EN_FILE = os.path.join(DATA_ROOT, "public_test.en.txt")

# Giá»›i háº¡n sá»‘ lÆ°á»£ng test samples (None = dÃ¹ng táº¥t cáº£)
MAX_TEST = None

# === Äá»c test set (VI lÃ  source, EN lÃ  reference) ===
with open(TEST_VI_FILE, "r", encoding="utf-8") as f:
    test_vietnamese = [line.rstrip("\n") for line in f]

# Kiá»ƒm tra xem cÃ³ file reference khÃ´ng
has_reference = os.path.exists(TEST_EN_FILE)
test_english = []
if has_reference:
    with open(TEST_EN_FILE, "r", encoding="utf-8") as f:
        test_english = [line.rstrip("\n") for line in f]
    if len(test_vietnamese) != len(test_english):
        raise ValueError(
            f"Sá»‘ dÃ²ng test khÃ´ng khá»›p! VI: {len(test_vietnamese)}, EN: {len(test_english)}"
        )

test_count = len(test_vietnamese)
if MAX_TEST is not None and MAX_TEST > 0:
    test_count = min(test_count, MAX_TEST)
    test_vietnamese = test_vietnamese[:test_count]
    if has_reference:
        test_english = test_english[:test_count]

print(f"\nâœ“ Sá»‘ máº«u test: {test_count}")
if has_reference:
    print("âœ“ CÃ³ file reference Ä‘á»ƒ tÃ­nh BLEU")
else:
    print("âš  KhÃ´ng cÃ³ file reference, chá»‰ sinh predictions")

# Chuyá»ƒn model sang inference mode
device = next(model.parameters()).device
FastLanguageModel.for_inference(model)

def translate_batch_vi_to_en(batch_vietnamese, max_new_tokens=256):
    """Dá»‹ch má»™t batch cÃ¢u tiáº¿ng Viá»‡t sang tiáº¿ng Anh"""
    prompts_list = []
    for vi_text in batch_vietnamese:
        messages = [
            {"role": "system", "content": TRANSLATION_SYSTEM_PROMPT},
            {"role": "user", "content": vi_text},
        ]
        chat_prompt = tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=True,
            enable_thinking=False,  # Táº¯t thinking mode cho translation
        )
        prompts_list.append(chat_prompt)

    # Tokenize batch
    tokenized_inputs = tokenizer(
        prompts_list,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=MAX_SEQUENCE_LENGTH,
    ).to(device)

    input_ids = tokenized_inputs["input_ids"]
    attention_mask = tokenized_inputs["attention_mask"]
    actual_lengths = attention_mask.sum(dim=1)  # Äá»™ dÃ i thá»±c cá»§a má»—i cÃ¢u (khÃ´ng tÃ­nh padding)

    # Generate translations
    with torch.no_grad():
        generated_outputs = model.generate(
            **tokenized_inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,  # Greedy decoding
        )

    # Extract translations (bá» pháº§n prompt)
    translations = []
    for i in range(generated_outputs.size(0)):
        generated_part = generated_outputs[i, actual_lengths[i]:]
        translated_text = tokenizer.decode(generated_part, skip_special_tokens=True).strip()
        translations.append(translated_text)
    
    return translations

# === Sinh predictions trÃªn test set ===
INFERENCE_BATCH_SIZE = 4  # CÃ³ thá»ƒ tÄƒng lÃªn 8/16 náº¿u GPU Ä‘á»§ máº¡nh

predictions = []
references = []
for start_idx in tqdm(range(0, test_count, INFERENCE_BATCH_SIZE), desc="Dá»‹ch test set (VI -> EN)"):
    end_idx = min(start_idx + INFERENCE_BATCH_SIZE, test_count)
    batch_vi = test_vietnamese[start_idx:end_idx]
    batch_ref = test_english[start_idx:end_idx] if has_reference else []

    batch_predictions = translate_batch_vi_to_en(batch_vi)
    predictions.extend(batch_predictions)
    if has_reference:
        references.extend(batch_ref)

# Hiá»ƒn thá»‹ 3 vÃ­ dá»¥ Ä‘áº§u
print("\n" + "="*60)
print("VÃ Dá»¤ 3 CÃ‚U Äáº¦U (VI -> EN)")
print("="*60)
for i in range(min(3, test_count)):
    print(f"\n[Test {i+1}]")
    print(f"VI  : {test_vietnamese[i]}")
    if has_reference:
        print(f"REF : {references[i]}")
    print(f"PRED: {predictions[i]}")
    print("-" * 80)

# === TÃ­nh SacreBLEU (chá»‰ náº¿u cÃ³ reference) ===
if has_reference:
    bleu_metric = evaluate.load("sacrebleu")
    bleu_results = bleu_metric.compute(
        predictions=predictions,
        references=[[ref] for ref in references],  # SacreBLEU cáº§n format list[list[str]]
    )
    print("\n" + "="*60)
    print("Káº¾T QUáº¢ SACREBLEU (VI -> EN)")
    print("="*60)
    print(f"SacreBLEU score: {bleu_results['score']:.2f}")
    print(f"Chi tiáº¿t: {bleu_results}")
else:
    print("\nâš  KhÃ´ng cÃ³ reference Ä‘á»ƒ tÃ­nh BLEU, chá»‰ sinh predictions")
```

```python
# ============================================================================
# CELL 9: LÆ°u káº¿t quáº£ inference ra file (VI -> EN)
# ============================================================================

import pandas as pd

# 1) LÆ°u báº£ng Ä‘áº§y Ä‘á»§: cÃ¢u VI, REF EN (náº¿u cÃ³), PRED EN
output_data = {
    "vi": test_vietnamese,   # CÃ¢u nguá»“n tiáº¿ng Viá»‡t
    "pred_en": predictions,   # Model dá»‹ch ra tiáº¿ng Anh
}
if has_reference:
    output_data["ref_en"] = references  # Reference tiáº¿ng Anh (náº¿u cÃ³)

results_dataframe = pd.DataFrame(output_data)

os.makedirs(SAVE_DIR, exist_ok=True)

# LÆ°u CSV
csv_output_path = os.path.join(SAVE_DIR, "public_test_vi-en_predictions.csv")
results_dataframe.to_csv(csv_output_path, index=False, encoding="utf-8")
print(f"\nâœ“ ÄÃ£ lÆ°u báº£ng chi tiáº¿t vÃ o: {csv_output_path}")

# 2) LÆ°u file predictions riÃªng (.txt) - má»—i dÃ²ng lÃ  má»™t cÃ¢u dá»‹ch
predictions_file = os.path.join(SAVE_DIR, "public_test.pred.en.txt")
with open(predictions_file, "w", encoding="utf-8") as f:
    for pred_line in predictions:
        f.write(pred_line + "\n")
print(f"âœ“ ÄÃ£ lÆ°u file predictions vÃ o: {predictions_file}")

print("\n" + "="*60)
print("HOÃ€N Táº¤T (VI -> EN)")
print("="*60)
```
